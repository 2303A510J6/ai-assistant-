```python
def is_leap_year(year):
    if year % 400 == 0:
        return True
    if year % 100 == 0:
        return False
    return year % 4 == 0
```

Task 2: GCD of Two Numbers

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return abs(a)
```

Task 3: LCM of Two Numbers

```python
def lcm(a, b):
    if a == 0 or b == 0:
        return 0

    x, y = abs(a), abs(b)
    while y:
        x, y = y, x % y

    return abs(a * b) // x
```

Task 4: Binary to Decimal Conversion

```python
def binary_to_decimal(binary):
    if not all(bit in "01" for bit in binary):
        raise ValueError("Invalid binary number")

    decimal = 0
    for bit in binary:
        decimal = decimal * 2 + int(bit)
    return decimal
```

Task 5: Decimal to Binary Conversion

```python
def decimal_to_binary(n):
    if n == 0:
        return "0"

    negative = n < 0
    n = abs(n)
    binary = ""

    while n > 0:
        binary = str(n % 2) + binary
        n //= 2

    return "-" + binary if negative else binary
```

Task 6: Harshad (Niven) Number Check

```python
def is_harshad(n):
    if n <= 0:
        return False

    digit_sum = sum(int(d) for d in str(n))
    return n % digit_sum == 0
```
