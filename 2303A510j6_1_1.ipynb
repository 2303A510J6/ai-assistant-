Question 1: Zero-Shot Prompting – Leap Year Check

```python
def is_leap_year(year):
    if year % 4 == 0:
        return True
    else:
        return False
```

Question 2: One-Shot Prompting – GCD of Two Numbers

```python
def gcd(a, b):
    for i in range(min(a, b), 0, -1):
        if a % i == 0 and b % i == 0:
            return i
```

Question 3: Few-Shot Prompting – LCM Calculation

```python
def lcm(a, b):
    def gcd(x, y):
        while y:
            x, y = y, x % y
        return x
    return abs(a * b) // gcd(a, b)
```

Question 4: Zero-Shot Prompting – Binary to Decimal

```python
def binary_to_decimal(binary):
    return int(binary, 2)
```

Question 5: One-Shot Prompting – Decimal to Binary

```python
def decimal_to_binary(n):
    if n < 0:
        return "Negative numbers not supported"
    return bin(n)[2:]
```

Question 6: Few-Shot Prompting – Harshad (Niven) Number

```python
def is_harshad(n):
    if n == 0:
        return "Not a Harshad Number"
    digit_sum = sum(int(d) for d in str(n))
    return "Harshad Number" if n % digit_sum == 0 else "Not a Harshad Number"
```
